```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

```python
# 定义环境：0=空地，1=金币（+10），2=陷阱（-5），3=智能体起点
world = np.array([
    [0, 0, 2, 0, 0],
    [0, 2, 0, 0, 0],
    [0, 0, 0, 2, 0],
    [0, 0, 0, 0, 1],
    [3, 0, 0, 0, 0]
])

# 动作：上、下、左、右（用箭头方向表示）
actions = [(-1,0), (1,0), (0,-1), (0,1)]  # 对应↑、↓、←、→
```

```python
def draw_world(pos, episode, total_reward, path=[]):
    plt.figure(figsize=(6,6))
    # 画格子背景
    for i in range(5):
        for j in range(5):
            if world[i,j] == 1:  # 金币
                plt.scatter(j, i, c='yellow', s=500, marker='o', edgecolors='black')
            elif world[i,j] == 2:  # 陷阱
                plt.scatter(j, i, c='white', s=500, marker='X', edgecolors='red', linewidths=3)
            elif world[i,j] == 3:  # 起点
                plt.scatter(j, i, c='green', s=500, marker='s', edgecolors='black')
    # 画智能体当前位置
    plt.scatter(pos[1], pos[0], c='blue', s=300, marker='s', edgecolors='black')
    # 画历史路径（浅灰色轨迹）
    if path:
        path_x = [p[1] for p in path]
        path_y = [p[0] for p in path]
        plt.plot(path_x, path_y, 'gray', linewidth=2, alpha=0.5)
    # 显示训练信息
    plt.title(f"train epoch:{episode} | total award:{total_reward}", fontsize=12)
    plt.xticks(range(5))
    plt.yticks(range(5))
    plt.grid(True, linewidth=2)
    plt.show()

# 显示初始状态（智能体在起点）
start_pos = tuple(np.argwhere(world == 3)[0])  # (4,0)
draw_world(start_pos, episode=0, total_reward=0)
```

```python
def draw_world(pos, episode, total_reward, path=[]):
    plt.figure(figsize=(6,6))
    # 画格子背景
    for i in range(5):
        for j in range(5):
            if world[i,j] == 1:  # 金币
                plt.scatter(j, i, c='yellow', s=500, marker='o', edgecolors='black')
            elif world[i,j] == 2:  # 陷阱
                plt.scatter(j, i, c='white', s=500, marker='X', edgecolors='red', linewidths=3)
            elif world[i,j] == 3:  # 起点
                plt.scatter(j, i, c='green', s=500, marker='s', edgecolors='black')
    # 画智能体当前位置
    plt.scatter(pos[1], pos[0], c='blue', s=300, marker='s', edgecolors='black')
    # 画历史路径（浅灰色轨迹）
    if path:
        path_x = [p[1] for p in path]
        path_y = [p[0] for p in path]
        plt.plot(path_x, path_y, 'gray', linewidth=2, alpha=0.5)
    # 显示训练信息
    plt.title(f"train epoch:{episode} | total award:{total_reward}", fontsize=12)
    plt.xticks(range(5))
    plt.yticks(range(5))
    plt.grid(True, linewidth=2)
    plt.show()

# 显示初始状态（智能体在起点）
start_pos = tuple(np.argwhere(world == 3)[0])  # (4,0)
draw_world(start_pos, episode=0, total_reward=0)
```

```python
# Q表：25个格子 × 4个动作，初始全为0（AI一开始啥都不懂）
q_table = np.zeros((25, 4))  # 25=5×5格子，4=四个方向动作
```

```python
def pos_to_idx(pos):
    return pos[0] * 5 + pos[1]

# Q-learning 参数
epsilon = 0.1  # 探索率
alpha = 0.1    # 学习率
gamma = 0.9    # 折扣因子
num_episodes = 100  # 训练回合数

# 训练过程
for episode in range(num_episodes):
    current_pos = start_pos  # 从起点开始
    total_reward = 0
    path = [current_pos]
    done = False

    while not done:
        current_state_idx = pos_to_idx(current_pos)

        # ε-greedy 策略选择动作
        if np.random.uniform() < epsilon:
            action = np.random.choice(4)  # 探索：随机选择一个动作
        else:
            action = np.argmax(q_table[current_state_idx, :])  # 利用：选择Q值最大的动作

        # 执行动作，计算新的位置和奖励
        move = actions[action]
        new_pos = (current_pos[0] + move[0], current_pos[1] + move[1])

        # 检查新位置是否越界
        if new_pos[0] < 0 or new_pos[0] >= 5 or new_pos[1] < 0 or new_pos[1] >= 5:
            new_pos = current_pos  # 越界则留在原地
            reward = -1  # 越界惩罚
            done = False # 越界不算结束回合
        else:
            # 根据环境获取奖励
            if world[new_pos] == 1:
                reward = 10  # 金币奖励
                done = True  # 找到金币结束回合
            elif world[new_pos] == 2:
                reward = -5  # 陷阱惩罚
                done = True  # 掉入陷阱结束回合
            else:
                reward = -0.1 # 空地微小惩罚
                done = False # 在空地不结束回合

        # 更新Q表
        new_state_idx = pos_to_idx(new_pos)
        best_next_q = np.max(q_table[new_state_idx, :])
        q_table[current_state_idx, action] = q_table[current_state_idx, action] + alpha * (reward + gamma * best_next_q - q_table[current_state_idx, action])

        current_pos = new_pos
        total_reward += reward
        path.append(current_pos)

        # 每隔一定回合数或在最后一个回合绘制世界状态
        if episode % 50 == 0 or episode == num_episodes - 1:
             clear_output(wait=True)
             draw_world(current_pos, episode, total_reward, path)
             time.sleep(0.1) # 添加延迟，方便观察

print("训练结束！")
```

```python
episodes = 50
learning_r = 0.1
discount_r = 0.9
epsilon = 0.2
def pos_o(pos):
  return pos[0] * 5 + pos[1]
for episode in range(episodes):
  current_pos = start_pos
  total_reward = 0
  path = [current_pos]
  done = False

  while not done:
    if np.random.random() < epsilon:
      action = np.random.choice(4)
    else:
      state_idx = pos_o(current_pos)
      action = np.argmax(q_table[state_idx])
    dx,dy = actions[action]
    new_pos = (current_pos[0] + dx, current_pos[1] + dy)
    if new_pos[0] < 0 or new_pos[0] >= 5 or new_pos[1] < 0 or new_pos[1] >= 5:
      new_pos = current_pos
      reward = -3
    else:
      if world[new_pos] == 1:
        reward = 10
        done = True
      elif world[new_pos] == 2:
        reward = -5
      else:
        reward = -1
    current_idx = pos_o(current_pos)
    new_idx = pos_o(new_pos)
    q_table[current_idx, action] += learning_r * (reward + discount_r * np.max(q_table[new_idx]) - q_table[current_idx, action])
    current_pos = new_pos
    total_reward += reward
    path.append(current_pos)

    clear_output(wait = True)
    draw_world(current_pos, episode, total_reward, path)
    time.sleep(0.3)
time.sleep(1)
```
